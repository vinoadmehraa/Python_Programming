```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_cleaned.csv')
```

```python
df
```

```python
df.drop(columns=['day','month','year'],inplace=True)
```

```python
df
```

```python
df['Classes'] = np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df
```

```python
## Divide the features in dependant and independant.
X = df.drop('FWI',axis=1)
y = df['FWI']
```

```python
X
```

```python
y
```

```python
## Train Test split.
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=20, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
## Feature selection based on correlation.

X_train.corr()
```

```python
## check for multicollinerality.
plt.figure(figsize=(15,5))
corr=X_train.corr()
sns.heatmap(corr,annot=True)
plt.show()
```

```python
## define a function which will return the column which is above the threshold.

## The threshold would be defined by the SME or product manager.

def correlation(dataset,threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j] > threshold):
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
## Check which column is above the threshold - 90%
correlation(X_train, 0.90)
```

```python
## drop the features which are 85% correlated.
corr_feature_drop = correlation(X_train, 0.85)
corr_feature_drop
```

```python
plt.figure(figsize=(15,5))
sns.heatmap(X_train.corr(), annot=True)
plt.show()
```

```python
## dropping the features.
X_train.head(5)
```

```python
X_train.drop(corr_feature_drop, axis=1, inplace=True)
```

```python
X_train.head(5)
```

```python
X_test.drop(corr_feature_drop,axis=1, inplace=True)
```

```python
X_train.shape, X_test.shape
```

## Feature Scalling.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_test_scaled
```

## Box plot to understand the effect of standard scaler.

```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(X_train)
plt.title("X_train before scalling")

plt.subplot(1,2,2)
sns.boxplot(X_train_scaled)
plt.title("X_train after scalling")

plt.show()
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
lin_reg = LinearRegression()

lin_reg.fit(X_train_scaled, y_train)
y_pred = lin_reg.predict(X_test_scaled)

mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test, y_pred)
lin_reg_intercept = lin_reg.intercept_
lin_reg_coef = lin_reg.coef_

# Constructing the regression equation
feature_names = X_train.columns  # Assuming X_train_scaled is a DataFrame
lin_reg_equation = f"y = {lin_reg_intercept:.4f}"
for coef, feature in zip(lin_reg_coef, feature_names):
    lin_reg_equation += f" + ({coef:.4f} * {feature})"
```

```python
print("The mean absolute error is: ",mae)
print("The R2 score is: ",score)

print("The intercept of the regression is: ",lin_reg_intercept)
print("All the co-efficients of the regression are: ",lin_reg_coef)

print("Regression Equation:", lin_reg_equation)
```

```python
plt.scatter(y_pred, y_test)
```

## Apply the lasso regresion.

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error, r2_score


lasso_reg = Lasso()

lasso_reg.fit(X_train_scaled, y_train)
y_pred = lasso_reg.predict(X_test_scaled)

mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test, y_pred)

lasso_reg_intercept = lasso_reg.intercept_
lasso_reg_coef = lasso_reg.coef_

# Constructing the Lasso regression equation
feature_names = X_train.columns  # Assuming X_train_scaled is a DataFrame
lasso_reg_equation = f"y = {lasso_reg_intercept:.4f}"
for coef, feature in zip(lasso_reg_coef, feature_names):
    lasso_reg_equation += f" + ({coef:.4f} * {feature})"

print("The mean absolute error is: ",mae)
print("The R2 score is: ",score)

print("The intercept of the Lasso regression is: ",lasso_reg_intercept)
print("All the co-efficients of the regression are: ",lasso_reg_coef)

print("Lasso Regression Equation:", lasso_reg_equation)

plt.scatter(y_pred, y_test)
```
