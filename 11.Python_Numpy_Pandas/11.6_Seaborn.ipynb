```python
import seaborn as sns
```

```python
## Basic plotting with seaborn.

tips_data=sns.load_dataset('tips')

tips_data
```

```python
## Scatter plot.
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips_data)
plt.title('Scatter plot for total bills vs tips')
plt.show()
```

```python
## line plot.

sns.lineplot(x='size',y='total_bill',data=tips_data)
plt.title('Line plot for total bills vs size')
plt.show()
```

```python
## Categorical variable Plot.

sns.barplot(x='day',y='total_bill',data=tips_data)
plt.title('Bar plot for total bills vs days')
plt.show()

```

```python
## Box plot.

sns.boxplot(x='day',y='total_bill', data=tips_data)
plt.title('box plot for total bills vs size')
plt.show()
```

```python
## Violine plot.

sns.violinplot(x='day',y='total_bill',data=tips_data)
plt.title('Violin plot for total bills vs day')
plt.show()

```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
sns.boxplot(x='day',y='total_bill', data=tips_data)
plt.title('box plot for total bills vs size')

plt.subplot(1,2,2)
sns.violinplot(x='day',y='total_bill',data=tips_data)
plt.title('Violin plot for total bills vs size')
plt.show()
```

```python
## Histogram plot.
sns.histplot(tips_data['total_bill'],bins=10, kde=True)
```

```python
## KDE - Kernel Density Plot.

sns.kdeplot(tips_data['total_bill'],fill=True)
```

```python
## Pair plot.

sns.pairplot(tips_data)
```

```python
## Correlation.

## Heatmap

my_corr = tips_data[['total_bill','tip','size']].corr()
my_corr
```

```python
sns.heatmap(my_corr,annot=True, cmap='coolwarm')
```

```python
## Customization of plot.

import pandas as pd
import numpy as np
```

```python
sale_df=pd.read_csv('sales_data.csv')

sale_df.head(5)
```

```python
## plot total sales by product.

plt.figure(figsize=(10,5))
sns.barplot(x='Product Category',y='Total Revenue', data=sale_df, estimator=sum)
plt.title('Total sales by product')
plt.xlabel('Product_Category')
plt.ylabel('Total_Revenue')
plt.show()
```

```python
sale_df
```

```python
sale_df_sorted = sale_df.groupby('Product Category')['Total Revenue'].sum().reset_index()

sale_df_sorted
```

```python
sale_df_sorted_tr = sale_df_sorted.sort_values(by='Total Revenue')

sale_df_sorted_tr
```

```python
# Plot of Product Category vs total revenue.

plt.figure(figsize=(10,5))
sns.barplot(x='Product Category',y='Total Revenue', data=sale_df_sorted_tr, estimator=sum)
plt.title('Total sales by product')
plt.xlabel('Product_Category')
plt.ylabel('Total_Revenue')
plt.show()
```

```python
# Plot of Product Category vs total revenue.

sale_df_sorted_pc = sale_df_sorted.sort_values(by='Product Category')

plt.figure(figsize=(10,5))
sns.barplot(x='Product Category',y='Total Revenue', data=sale_df_sorted_pc, estimator=sum, hue='Product Category')
plt.title('Total sales by product')
plt.xlabel('Product_Category')
plt.ylabel('Total_Revenue')

plt.savefig('pc_revenue.png', dpi=30, transparent=True, bbox_inches='tight')

plt.show()
```
