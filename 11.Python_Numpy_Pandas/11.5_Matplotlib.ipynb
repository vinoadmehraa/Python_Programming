```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]

y=[1,2,4,5,7,6]

## Create a line plot.

plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Basic Line plot')
plt.show()
```

```python
## Customized plot.
x=[1,2,3,4,5,6]
y=[1,1.5,4.5,5,5.5,7]

## linestyle: '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'

plt.plot(x,y, color='r', linestyle='dotted', marker='o',  linewidth=5, markersize='9') 
plt.grid(True)
```

```python
## Multiple plot.

x=[1,2,3,4,5,6]
y1=[1,1.5,4.5,5,5.5,7]
y2=[1,2,3,4,5,6]

plt.figure(figsize=(15,9))


plt.subplot(2,2,1) ## Row, columns, which cell position
plt.plot(x,y1,color='blue')
plt.title('First plot of X and Y1')

plt.subplot(2,2,4) ## Row, columns, which cell.
plt.plot(x,y2,color='red')
plt.title('First plot of X and Y2')

plt.subplot(2,2,2) ## Row, columns, which cell position
plt.plot(x,y2,color='green')
plt.title('First plot of X and Y2')

plt.subplot(2,2,3) ## Row, columns, which cell.
plt.plot(x,y1,color='yellow')
plt.title('First plot of X and Y1')

```

```python
## Bar plots.

categories=['A','B','C','D','E']
values=[1,2,3,4,5]

plt.bar(categories, values,color='purple')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Basic Bar plot')
plt.show()
```

```python
## Histogram.

data=[1,2,3,4,2,3,4,5,6,2,3,5,6,7,8,9,4,5,6,7,8,9]

plt.hist(data,bins=5, color='orange', edgecolor='black')
```

```python
## Creating a scatter plot.

x=[1,2,3,4]
y=[1,2.5,3.1,5]

plt.scatter(x,y,color='blue',marker='x')
```

```python
## Pie Chart

labels=['A','B','C','D']
sizes=[10,20,30,40]
my_color=['gold','yellowgreen','lightcoral','lightskyblue']
my_explode = (0.2,0,0,0) ## Move out the first slice, moveout 2nd slice, moveout 3rd slice, moveout 4th slice.

plt.pie(sizes,labels=labels, colors=my_color, autopct="%1.1f%%", shadow=True, explode=my_explode)
```

```python
## Practical examples.

import pandas as pd

sale_date=pd.read_csv('sales_data.csv')

sale_date.head(5)
```

```python
## plot total sales by products.

sale_date.info()
```

```python
total_sales_by_product = sale_date.groupby('Product Category')['Total Revenue'].sum()

total_sales_by_product
```

```python
## plot the bar chart.

total_sales_by_product.plot(kind='bar', color='teal')
```

```python
## plot sales over time.

sales_trend = sale_date.groupby('Date')['Total Revenue'].sum()

sales_trend
```

```python
sales_trend = sale_date.groupby('Date')['Total Revenue'].sum().reset_index()

sales_trend
```

```python
plt.plot(sales_trend['Date'],sales_trend['Total Revenue'], color='teal')
```

```python
sales_trend.plot(kind='line', color='teal')
```
